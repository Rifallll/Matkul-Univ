4.D. Buat **fungsi** untuk mengkategorikan status gizi dengan syarat:

  jika:
  * berat_badan dibawah -3 SD = BB Sangat Kurang
  * berat badan -3 SD  hingga -2 SD = BB Kurang
  * berat_badan -2 SD hingga +1 SD = BB Normal
  * diatas +2 SD = Berat Badan Lebih

````python
#4.D.
def categorize_nutritional_status(age_in_months, weight, df):
    if age_in_months < 0 or age_in_months > 24:
        return "Age out of range (0-24 months)"

    weights_at_age = df.loc[age_in_months]
    if weight < weights_at_age['-3 SD']:
        return "BB Sangat Kurang"
    elif weight >= weights_at_age['-3 SD'] and weight < weights_at_age['-2 SD']:
        return "BB Kurang"
    elif weight >= weights_at_age['-2 SD'] and weight < weights_at_age['+2 SD']:
        return "BB Normal"
    elif weight >= weights_at_age['+2 SD']:
        return "Berat Badan Lebih"
    else:
        return "Could not determine status"
        print (weights_at_age)

````

4.E hubungkan fungsi pada 4.D dengan inputan pada nomor 4.C, input umur dan berat badan, output status gizi + plot

````python
try:
    age_in_months = int(input("Masukkan umur dalam bulan (0-24): "))
    weight = float(input("Masukkan berat badan (kg): "))

    def categorize_nutritional_status(age_in_months, weight, df):
        if age_in_months < 0 or age_in_months > 24:
            return "Age out of range (0-24 months)"

        weights_at_age = df.loc[age_in_months]
        if weight < weights_at_age['-3 SD']:
            return "BB Sangat Kurang"
        elif weight >= weights_at_age['-3 SD'] and weight < weights_at_age['-2 SD']:
            return "BB Kurang"
        elif weight >= weights_at_age['-2 SD'] and weight < weights_at_age['+2 SD']:
            return "BB Normal"
        elif weight >= weights_at_age['+2 SD']:
            return "Berat Badan Lebih"
        else:
            return "Could not determine status"
            print (weights_at_age)

    nutritional_status = categorize_nutritional_status(age_in_months, weight, df)

    print("\n----- HASIL ANALISIS -----")
    print(f"Umur: {age_in_months} bulan")
    print(f"Berat Badan: {weight} kg")
    print(f"Status Gizi: {nutritional_status}")
    print("--------------------------\n")

    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(12, 8))

    ax.plot(df.index, df['Median'], 'g-', label='Median', linewidth=2)
    ax.plot(df.index, df['-2 SD'], 'y--', label='-2 SD (Batas Bawah Normal)')
    ax.plot(df.index, df['+2 SD'], 'y--', label='+2 SD (Batas Atas Normal)')
    ax.plot(df.index, df['-3 SD'], 'r--', label='-3 SD (Gizi Sangat Kurang)')

    ax.fill_between(df.index, df['-2 SD'], df['+2 SD'], color='green', alpha=0.1, label='Rentang Normal')
    ax.plot(age_in_months, weight, 'r*', markersize=15, label=f'Posisi Anda: {nutritional_status}')

    ax.set_title('Grafik Pertumbuhan Berat Badan vs Umur (0-24 Bulan)', fontsize=16)
    ax.set_xlabel('Umur (Bulan)', fontsize=12)
    ax.set_ylabel('Berat Badan (kg)', fontsize=12)
    ax.set_xticks(np.arange(0, 25, 1))
    ax.legend()
    plt.grid(True)

    plt.savefig("grafik_pertumbuhan_anak.png")
    print("Grafik telah disimpan sebagai 'grafik_pertumbuhan_anak.png'")

except ValueError:
    print("Input tidak valid. Harap masukkan angka yang benar.")
except Exception as e:
    print(f"Terjadi error: {e}")
````
